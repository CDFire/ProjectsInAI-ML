# Task 1

**1.** In logistic regression, the goal is to find  P(y | x)  for binary classification of y. The equation for which is as follows:

    P(y = 1 | x) = h_θ(x) = σ(x^⊤ θ)

The probability of y = 0  is:

    P(y = 0 | x) = 1 - σ(x^⊤ θ)

The likelihood function quantifies how likely the observed data is given the parameters of the model:

    L(θ) = ∏ᵢⁿP(yᵢ | xᵢ, θ)

We can substitute the probabily equations above into the likelyhood function to get:

    L(θ) = ∏ᵢⁿ (h_θ(xᵢ))^yᵢ (1 - h_θ(xᵢ))^(1 - yᵢ)

To simplify computations, we then take the logarithm of the likelihood:

    l(θ) = log(L(θ)) = ∑ᵢⁿ (yᵢ log(h_θ(xᵢ)) + (1 - yᵢ)log(1 - h_θ(xᵢ))

We then use MLE to get the objective function from the log likelyhood. We do this by treating the log-likelihood as a maximization problem. However, we actually need to do this as a minimization instead of a maximization problem so we simply use the negative of the log-likelihood. This means that the objective function is simply minimizing the negative log-likelihood.

MAP incorporates prior knowledge about the parameters θ into the learning process, extending MLE using Bayes’ theorem. By implementing regularization based on a prior distribution, MAP helps prevent overfitting.

Logistic Regression is inherently designed for binary classification tasks. In this scenario, where the goal is to classify employees into two distinct categories (at risk vs. not at risk of burnout), logistic regression provides a straightforward and effective solution. In addition, Logistic Regression outputs probabilities associated with each class. This probabilistic approach allows organizations to gauge the confidence level of each prediction. For instance, an employee with a 90% probability of burnout can be prioritized for immediate intervention compared to someone with a 60% probability. Logistic Regression also makes fewer assumptions about the underlying data distribution compared to other models. It doesn’t require features to be normally distributed or to have a linear relationship with the target variable, offering flexibility in diverse organizational data scenarios such as this.

Comparison with Support Vector Machine (SVM):
While Support Vector Machines are also powerful linear classification models, there are distinct differences when compared to Logistic Regression in the context of predicting employee burnout. SVMs focus on finding the optimal hyperplane that maximizes the margin between two classes while logistic regression aims to model the probability of class membership. This provides a smooth probability distribution rather than a hard decision boundary which is better for predicting burnout as you want to see how at risk an individual is instead of a simple binary yes or no. In addition, while SVMs with linear kernels are interpretable to some extent, the focus on margin maximization can obscure the direct impact of each feature on the prediction. Logistic Regression offers clear interpretability where each coefficient directly relates to the log-odds of the target variable, making it easier for stakeholders to understand feature influences which is very important in understanding what to change to decrease burnout among employees.

Each of these features would need to be encoded into some numerical representation, but once they are, each of these features would constitute an element of the feature vector xᵢ for the  i-th employee:


xᵢ = \begin{bmatrix}
\text{Sick Days}ᵢ \\
\text{Late Arrivals}ᵢ \\
\text{Sentiment Score}ᵢ \\
\vdots \\
\text{Team Dynamics}ᵢ

The parameter vector θ contains the weights corresponding to each feature in  x. These weights quantify the influence of each feature on the probability of an employee being at risk of burnout.

The target variable yᵢ is a binary indicator for each employee:

yᵢ = \begin{cases}
1 & \text{if employee } i \text{ is at risk of burnout} \\
0 & \text{otherwise}
\end{cases}

h_Θ(xᵢ) represents the probability that employee i is at risk of burnout.

The likelihood function aggregates the probabilities of observing the actual burnout statuses across all employees in the dataset, given weights Θ.

The log-likelihood simplifies the computation by converting the product of probabilities into a sum, facilitating easier differentiation during optimization.

The objective function is minimized to find the optimal weights θ that maximize the likelihood of observing the data.

Assumptions:

The derivation of the logistic regression objective function using MLE makes a few assumptions about the dataset and the underlying data-generating process:

•	Each employee’s burnout status yᵢ is independent of others, given their feature vector xᵢ. This is because the likelihood function is the product of individual probabilities. This simplifies computation but may not hold in cases where employee burnout is influenced by group dynamics or interactions.

•	The log-odds of the probability of burnout are a linear combination of the features. This may not capture complex, nonlinear interactions between features.

• The sample size is sufficiently large to provide reliable estimates. As Logistic Regression requires adequate data to estimate the parameters accurately, this may be difficult to get in corporate settings depending on employee numbers.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
class LogisticRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000, gradient_descent_type='batch', batch_size=64):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.gradient_descent_type = gradient_descent_type
        self.batch_size = batch_size
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        #compute the sigmoid of z
        return 1 / (1 + np.exp(-z))

    def initialize_parameters(self, n_features):
        #initialize weights and bias
        self.weights = np.zeros((n_features, 1))
        self.bias = 0.0

    def compute_cost(self, y_true, y_pred):
        #compute the cost function
        m = y_true.shape[0]
        epsilon = 1e-15  # To prevent log(0)
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        cost = -(1 / m) * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
        return cost

    def compute_gradients(self, x, y_true, y_pred):
        #compute the gradients of weights and bias
        m = x.shape[0]
        dw = (1 / m) * np.dot(x.T, (y_pred - y_true))
        db = (1 / m) * np.sum(y_pred - y_true)
        return dw, db

    def fit(self, X, y):
        #train the Logistic Regression model using the specified gradient descent method
        y = y.reshape(-1, 1)
        n_features = X.shape[1]
        self.initialize_parameters(n_features)

        for i in range(self.num_iterations):
            if self.gradient_descent_type == 'batch':
                #batch gradient descent
                z = np.dot(X, self.weights) + self.bias
                y_pred = self.sigmoid(z)
                cost = self.compute_cost(y, y_pred)
                dw, db = self.compute_gradients(X, y, y_pred)

                self.weights -= self.learning_rate * dw
                self.bias -= self.learning_rate * db

                #compute cost after each epoch
                z = np.dot(X, self.weights) + self.bias
                y_pred = self.sigmoid(z)
                cost = self.compute_cost(y, y_pred)

                if i % 50 == 0:
                  print(f"Iteration {i+1}/{self.num_iterations} - Cost: {cost:.4f}")

            elif self.gradient_descent_type == 'stochastic':
                #stochastic gradient descent
                for j in range(X.shape[0]):
                    xi = X[j].reshape(1, -1)
                    yi = y[j].reshape(1, 1)
                    z = np.dot(xi, self.weights) + self.bias
                    y_pred = self.sigmoid(z)
                    dw, db = self.compute_gradients(xi, yi, y_pred)

                    self.weights -= self.learning_rate * dw
                    self.bias -= self.learning_rate * db

                #compute cost after each epoch
                z = np.dot(X, self.weights) + self.bias
                y_pred = self.sigmoid(z)
                cost = self.compute_cost(y, y_pred)

                if i % 50 == 0:
                  print(f"Iteration {i+1}/{self.num_iterations} - Cost: {cost:.4f}")

            elif self.gradient_descent_type == 'mini-batch':
                #mini-batch gradient descent
                indices = np.random.permutation(X.shape[0])
                X_shuffled = X[indices]
                y_shuffled = y[indices]
                for j in range(0, X.shape[0], self.batch_size):
                    X_mini = X_shuffled[j:j+self.batch_size]
                    y_mini = y_shuffled[j:j+self.batch_size]
                    z = np.dot(X_mini, self.weights) + self.bias
                    y_pred = self.sigmoid(z)
                    dw, db = self.compute_gradients(X_mini, y_mini, y_pred)

                    self.weights -= self.learning_rate * dw
                    self.bias -= self.learning_rate * db

                #compute cost after each epoch
                z = np.dot(X, self.weights) + self.bias
                y_pred = self.sigmoid(z)
                cost = self.compute_cost(y, y_pred)

                if i % 50 == 0:
                  print(f"Iteration {i+1}/{self.num_iterations} - Cost: {cost:.4f}")
        return self

    def predict(self, X):
        return (self.sigmoid(np.dot(X, self.weights) + self.bias) > 0.5).astype(int)
```

```python
x, y = make_classification(n_samples=3000, n_features=8, n_classes=2)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

#compare gradient descent variants
for gd_type in ['batch', 'stochastic', 'mini-batch']:
    print(f"\nUsing {gd_type} gradient descent:")
    #use gradient descent type to train the model
    model = LogisticRegression(learning_rate=0.01, num_iterations=300, gradient_descent_type=gd_type, batch_size=32)
    model.fit(x_train, y_train)

    #evaluate the model
    y_pred = model.predict(x_test)
    accuracy = np.mean(y_pred.flatten() == y_test) * 100
    print(f"Accuracy: {accuracy:.2f}%")
```

1.	Batch gradient descent computes the gradient of the cost function using the entire training dataset in each iteration. Since it uses all data points, the gradient estimate is accurate, leading to smooth and stable convergence. We see this above as the cost consistently decreases across iterations, showcasing stable convergence.

2.	Stochastic gradient descent computes the gradient using only a single randomly selected training example per iteration. The inherent noise can help escape shallow local minima, potentially finding better solutions. This resulted in a rapid initial decrease as we see the cost dropped quickly fromwithin the first few iterations but remained constant afterwards.

3.	Mini-Batch gradient descent omputes the gradient using a small, randomly selected subset of the training dataset per iteration. Mini-Batch is a solid compromise between SGD and BGD. Mini-batch usually converges faster than batch methods due to more frequent updates. We see this in the big initial decrease over the first 51 iterations. Similar to SGD, the cost stabilizes around and plateus after few iterations.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
class LogisticRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000, gradient_descent_type='batch', batch_size=64, optimizer='sgd', beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.gradient_descent_type = gradient_descent_type
        self.batch_size = batch_size
        self.optimizer = optimizer.lower()
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.weights = None
        self.bias = None
        self.v_dw = None
        self.v_db = None
        self.s_dw = None
        self.s_db = None
        self.m_dw = None
        self.m_db = None
        self.v_m_dw = None
        self.v_m_db = None
        self.v_v_dw = None
        self.v_v_db = None
        self.iteration = 0

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def initialize_parameters(self, n_features):
        self.weights = np.zeros((n_features, 1))
        self.bias = 0.0

        if self.optimizer in ['momentum', 'adam']:
            self.v_dw = np.zeros_like(self.weights)
            self.v_db = 0.0
        if self.optimizer in ['rmsprop', 'adam']:
            self.s_dw = np.zeros_like(self.weights)
            self.s_db = 0.0
        if self.optimizer == 'adam':
            self.m_dw = np.zeros_like(self.weights)
            self.m_db = 0.0
            self.v_v_dw = np.zeros_like(self.weights)
            self.v_v_db = 0.0

    def compute_cost(self, y_true, y_pred):
        m = y_true.shape[0]
        epsilon = 1e-15
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        cost = -(1 / m) * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
        return cost

    def compute_gradients(self, x, y_true, y_pred):
        m = x.shape[0]
        dw = (1 / m) * np.dot(x.T, (y_pred - y_true))
        db = (1 / m) * np.sum(y_pred - y_true)
        return dw, db

    def fit(self, X, y):
        y = y.reshape(-1, 1)
        n_features = X.shape[1]
        self.initialize_parameters(n_features)

        for i in range(1, self.num_iterations + 1):
            if self.gradient_descent_type == 'batch':
                z = np.dot(X, self.weights) + self.bias
                y_pred = self.sigmoid(z)
                cost = self.compute_cost(y, y_pred)
                dw, db = self.compute_gradients(X, y, y_pred)

                self.update_parameters(dw, db)

            elif self.gradient_descent_type == 'stochastic':
                for j in range(X.shape[0]):
                    xi = X[j].reshape(1, -1)
                    yi = y[j].reshape(1, 1)
                    z = np.dot(xi, self.weights) + self.bias
                    y_pred = self.sigmoid(z)
                    dw, db = self.compute_gradients(xi, yi, y_pred)

                    self.update_parameters(dw, db)

            elif self.gradient_descent_type == 'mini-batch':
                indices = np.random.permutation(X.shape[0])
                X_shuffled = X[indices]
                y_shuffled = y[indices]
                for j in range(0, X.shape[0], self.batch_size):
                    X_mini = X_shuffled[j:j+self.batch_size]
                    y_mini = y_shuffled[j:j+self.batch_size]
                    z = np.dot(X_mini, self.weights) + self.bias
                    y_pred = self.sigmoid(z)
                    dw, db = self.compute_gradients(X_mini, y_mini, y_pred)

                    self.update_parameters(dw, db)
        return self

    def update_parameters(self, dw, db):
        if self.optimizer == 'sgd':
            #vanilla sgd
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

        elif self.optimizer == 'momentum':
            #momentum
            self.v_dw = self.beta1 * self.v_dw + (1 - self.beta1) * dw
            self.v_db = self.beta1 * self.v_db + (1 - self.beta1) * db

            self.weights -= self.learning_rate * self.v_dw
            self.bias -= self.learning_rate * self.v_db

        elif self.optimizer == 'rmsprop':
            #RMSProp
            self.s_dw = self.beta2 * self.s_dw + (1 - self.beta2) * (dw ** 2)
            self.s_db = self.beta2 * self.s_db + (1 - self.beta2) * (db ** 2)

            self.weights -= self.learning_rate * dw / (np.sqrt(self.s_dw) + self.epsilon)
            self.bias -= self.learning_rate * db / (np.sqrt(self.s_db) + self.epsilon)

        elif self.optimizer == 'adam':
            #adam
            self.iteration += 1
            self.m_dw = self.beta1 * self.m_dw + (1 - self.beta1) * dw
            self.m_db = self.beta1 * self.m_db + (1 - self.beta1) * db
            self.v_v_dw = self.beta2 * self.v_v_dw + (1 - self.beta2) * (dw ** 2)
            self.v_v_db = self.beta2 * self.v_v_db + (1 - self.beta2) * (db ** 2)

            m_dw_hat = self.m_dw / (1 - self.beta1 ** self.iteration)
            m_db_hat = self.m_db / (1 - self.beta1 ** self.iteration)
            v_dw_hat = self.v_v_dw / (1 - self.beta2 ** self.iteration)
            v_db_hat = self.v_v_db / (1 - self.beta2 ** self.iteration)

            self.weights -= self.learning_rate * m_dw_hat / (np.sqrt(v_dw_hat) + self.epsilon)
            self.bias -= self.learning_rate * m_db_hat / (np.sqrt(v_db_hat) + self.epsilon)

    def predict(self, X):
        return (self.sigmoid(np.dot(X, self.weights) + self.bias) > 0.5).astype(int)
```

```python
x, y = make_classification(n_samples=5000, n_features=8, n_classes=2)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
def train_and_evaluate(optimizer_name):
    print(f"\nTraining with {optimizer_name.capitalize()} Optimizer:")
    model = LogisticRegression(
        learning_rate=0.01,
        num_iterations=300,
        gradient_descent_type='mini-batch',
        batch_size=64,
        optimizer=optimizer_name
    )
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)

    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, predictions)
    precision = precision_score(y_test, predictions)
    recall = recall_score(y_test, predictions)
    f1 = f1_score(y_test, predictions)

    print(f"Accuracy with {optimizer_name.capitalize()}: {accuracy * 100:.2f}%")
    print(f"Precision with {optimizer_name.capitalize()}: {precision * 100:.2f}%")
    print(f"Recall with {optimizer_name.capitalize()}: {recall * 100:.2f}%")
    print(f"F1 Score with {optimizer_name.capitalize()}: {f1 * 100:.2f}%")

    # Return the model and metrics
    return model, {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1}
```

```python
optimizers = ['sgd', 'momentum', 'rmsprop', 'adam']
models = {}

for opt in optimizers:
    models[opt] = train_and_evaluate(opt)
```

```python
import itertools
param_grids = {
    'sgd': {
        'learning_rate': [0.001, 0.01, 0.1],
    },
    'momentum': {
        'learning_rate': [0.001, 0.01, 0.1],
        'beta1': [0.5, 0.7, 0.9, 0.95, 0.99]
    },
    'rmsprop': {
        'learning_rate': [0.001, 0.01, 0.1],
        'beta2': [0.8, 0.9, 0.95, 0.99],
        'epsilon': [1e-8, 1e-7, 1e-6]
    },
    'adam': {
        'learning_rate': [0.001, 0.01, 0.1],
        'beta1': [0.8, 0.9, 0.95],
        'beta2': [0.9, 0.999, 0.9999],
        'epsilon': [1e-8, 1e-7, 1e-6]
    }
}

def grid_search(optimizer_name, param_grid, x_train, y_train, x_test, y_test):
    print(f"\nPerforming Grid Search for {optimizer_name.capitalize()} Optimizer")
    keys = list(param_grid.keys())
    values = list(param_grid.values())
    combinations = list(itertools.product(*values))

    best_metrics = {'Accuracy': 0}
    best_params = {}

    results = []

    for combo in combinations:
        params = dict(zip(keys, combo))
        model = LogisticRegression(
            learning_rate=params.get('learning_rate', 0.01),
            num_iterations=300,
            gradient_descent_type='mini-batch',
            batch_size=64,
            optimizer=optimizer_name,
            beta1=params.get('beta1', 0.9),
            beta2=params.get('beta2', 0.999),
            epsilon=params.get('epsilon', 1e-8)
        )
        model.fit(x_train, y_train)
        predictions = model.predict(x_test)

        accuracy = accuracy_score(y_test, predictions)
        precision = precision_score(y_test, predictions)
        recall = recall_score(y_test, predictions)
        f1 = f1_score(y_test, predictions)

        results.append({
            **params,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1
        })

        if accuracy > best_metrics['Accuracy']:
            best_metrics = {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1}
            best_params = params

    results_df = pd.DataFrame(results)

    print(f"Best Parameters for {optimizer_name.capitalize()} Optimizer: {best_params}")
    print(f"Best Metrics: {best_metrics}")

    return results_df, best_params, best_metrics
```

```python
def perform_hyperparameter_tuning(optimizer_name, n_iter=20):
    param_grid = param_grids[optimizer_name]
    results_df, best_params, best_metrics = grid_search(optimizer_name, param_grid, x_train, y_train, x_test, y_test)
    return results_df, best_params, best_metrics

optimizers = ['sgd', 'momentum', 'rmsprop', 'adam']

tuning_results = {}
best_tuning_params = {}
best_tuning_metrics = {}

print("=== Grid Search ===")
for opt in optimizers:
    results_df, best_params, best_metrics = perform_hyperparameter_tuning(opt)
    tuning_results[opt] = results_df
    best_tuning_params[opt] = best_params
    best_tuning_metrics[opt] = best_metrics
```
